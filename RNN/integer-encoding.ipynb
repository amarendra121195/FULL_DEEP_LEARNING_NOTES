```python
import numpy as np
```

```python
docs =["go india",
       "india india",
       "hip hip hooray",
       "jitega bhai jitega india india",
       "bharat mata ki jai",
       "kohali kohali",
       "sachin sachin",
       "dhoni dhoni",
       "modi ji ke jai",
       "inquilab zindabad"
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(oov_token="amarendra")
 # out of vocabulary token for unknown words, it means if any word is not present in the training data, it will be replaced by this token
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.index_word
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
# above the all sequences are integer encoded representation of the text data, now we will pad these sequences to make them of same length, using pad_sequences from keras.preprocessing.sequence
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post')
```

```python
sequences
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
X_train = pad_sequences(X_train, padding='post')
X_test = pad_sequences(X_test, padding='post')
```

```python
len(X_train[24999])
```

```python
len(X_train[2])
```

```python
model = Sequential()

model.add(SimpleRNN(32,input_shape=(2494,1),return_sequences=False)) # return_sequences=False means we only want the output at the last time step, not the output at each time step(but sometimes we may want the output at each time step, for example in sequence to sequence models)
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
